```python
country_respondents = df['Country'].value_counts()
```

```python
country_respondents
```

```python
country_uses_python = country_grp['WantWorkLanguage'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

```python
python_df = pd.concat([country_respondents, country_uses_python], axis='columns', sort=False)
```

```python
python_df
```

```python
python_df.rename(columns={'Country' : 'NumRespondents', 'WantWorkLanguage' : 'NumKnowPython'}, inplace=True)
```

```python
python_df    
```

```python
python_df['PctKnowsPython'] = (python_df['NumKnowPython']/python_df['NumRespondents']) * 100
python_df
```

```python
python_df.sort_values(by='PctKnowsPython', ascending=False, inplace=True)
```
